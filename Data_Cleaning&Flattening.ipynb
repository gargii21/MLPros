```python
a=0
subjects_count = 0
subjects_dict = []
for elem in df['Subject_ID']:
    # a+=1
    # if a==3:
    #   break
    #im = 'ADNI_002_S_0413_MR_MPR____N3__Scaled_Br_20070216232854688_S14782_I40657.nii'
    #img = nib.load(elem).get_data()
    #img = np.asarray(img)

    subject_name = elem

    if subjects_count==0:
        subjects_count +=1
        subjects_dict.append(subject_name)
    else:
        if subject_name in subjects_dict:
            subjects_count += 0
            #subjects_dict.append(subject_name)
	    #continue
        else:
            subjects_count += 1
            subjects_dict.append(subject_name)
    if subjects_count>1072:
         break
np.shape(subjects_dict)
```

```python
df.head(30)
```

```python
n = len(subjects_dict)

dfV = pd.DataFrame({
    's_id': subjects_dict,
    'bl': [None]*n,
    'm06': [None]*n,
    'm12': [None]*n,
    'm18': [None]*n,
    'm24': [None]*n,
    'm36': [None]*n
    # Add more columns as needed
})
```

```python
# Set Subject_ID as index for easier lookup in dfV
dfV.set_index('s_id', inplace=True)

# Iterate through rows in df
for _, row in df.iterrows():
    sid = row['Subject_ID']     # subject ID
    visit = row['Visit']        # visit label (e.g., 'bl', 'm06')
    features = row['new_features']  # feature list

    # Update dfV only if the subject and visit exist
    if sid in dfV.index and visit in dfV.columns:
        dfV.at[sid, visit] = features

# Reset index if needed
dfV.reset_index(inplace=True)
```

```python
dfV.head(30)
```

```python
for index, row in dfV.iterrows():
    for col in dfV.columns:
        if row[col] is None:
            dfV.at[index, col] = np.full(66, np.nan)
```

```python
final_LIST=[]
for index, row in dfV.iterrows():
  sublist=[]
  for column in dfV.columns:
    if column=='s_id':
      continue
    for val in row[column]:
      sublist.append(val)
  final_LIST.append(sublist)

print(final_LIST)
```

```python
dff = pd.DataFrame(final_LIST)
```
